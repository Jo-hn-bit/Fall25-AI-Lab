```python
TASKS_FILE = "all_tasks.txt"






def taking_input():


    TASKS = input("Enter the task: ")
    TASKS_PRIORITY = input("Enter the priority(low, high or medium) : ")
    DUE_DATE = input("Enter due date for the task if there is one : ")
    
    task_status = "[ ]"



    with open(TASKS_FILE, "a") as file:

        file.write(f"{task_status} {TASKS} priority level: {TASKS_PRIORITY}, Due date is {DUE_DATE}\n")
    print("Task added")


taking_input()



def for_viewing():

    with open(TASKS_FILE, "r") as file:

                tasks_in_file = file.readlines()
                


                # Print a header for the list.
                print("\n--- Your To-Do List ---")
                    # Loop through each task and print it.
                for i, tasks_in_file in enumerate(tasks_in_file, 1):
                        # The `strip()` method removes any extra whitespace or newlines.
                    print(f"{i}. {tasks_in_file.strip()}")
                print("-----------------------")

view_input = input("What are you looking for? ")
if view_input.lower() == "view_tasks":
      for_viewing()














```

```python
TASKS_FILE = "all_tasks.txt"
tasks_list = []

while True:
    print("Your tasks right now:", tasks_list)
    what_to_do = input("Do you want to add, view, remove or quit? ").lower()

    if what_to_do == "add":
        newtask = input("Write YOUR task: ")
        priority = input("What is the priority level (low, medium, high): ")
        tasks_list.append(newtask)
        print("task was added")
        with open(TASKS_FILE, "a") as tasks:

            tasks.write(f"{newtask} priority level: {priority}\n")
            print("Task added")

    elif what_to_do == "remove":
        taskname = input("Which task to remove? write exact: ")
        if taskname in tasks_list:
            tasks_list.remove(taskname)
            print("removed")
        else:
            print("task not found")
    
    elif what_to_do == "view tasks":
        tasks = tasks.read()
        print(tasks)

    elif what_to_do == "quit":
        print("ended")
        break

    else:
        print("wrong answer")




```
